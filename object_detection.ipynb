### train,valid,test의 각 이미지 리스트 정리

```python
img_list = glob('./cnv_total-2/train/images/*.jpg')
print(len(img_list))
```

```python
from sklearn.model_selection import train_test_split
train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))
```

```python
with open('./cnv_total-2/train.txt','w') as f:
    f.write ('\n'.join(train_img_list)+'\n')
```

```python
with open('./cnv_total-2/val.txt','w') as f:
    f.write ('\n'.join(val_img_list)+'\n')
```

```python
import yaml
with open('./cnv_total-2/data.yaml','r') as f:
    data=yaml.load(f, Loader=yaml.FullLoader)
    
data['train']='../cnv_total-2/train.txt'
data['val']='../cnv_total-2/val.txt'

with open('./cnv_total-2/data.yaml','w') as f:
    yaml.dump(data,f)
```
